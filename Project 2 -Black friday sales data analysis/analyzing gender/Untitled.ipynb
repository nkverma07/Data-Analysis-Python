```python
import pandas as pd
```

```python
df  = pd.read_csv('BlackFriday.csv')
```

```python
df.head()
```

```python
data = pd.DataFrame({'Ratio': [len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])]}, index = ['Male','Female'])
```

```python
data
```

```python
data.plot.pie(y = 'Ratio',subplots = True,autopct = '%.1f%%',figsize=(6,6))
```

```python
df.groupby('Gender').size().plot(kind='pie',figsize = (6,6),autopct = '%.1f%%',title = 'Gender Ratio')
```

```python
df.groupby('Gender').size().plot(kind = 'bar')
```
