```python
for i in range(1,len(data)):
    print(data[i][2],round(((data[i][3]-data[i-1][3])/data[i-1][3])*100,2))
```

```python
gdp_change = [0]
for i in range(1,len(data)):
    cur = data[i][3]
    prev = data[i-1][3]
    gdp_change.append(round(((cur-prev)/prev)*100,2))
df_pr.assign(GDP_Growth = gdp_change)
```
