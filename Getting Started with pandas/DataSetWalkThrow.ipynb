```python
data = data.values
```

# Filtering Category With ART_AND_DESIGN

```python
filter_data = data[data['Category']=='ART_AND_DESIGN']
filter_data
```

# Accessing column inside ART_AND_DESIGN

```python
review_size = filter_data.loc[:,['Reviews','Size']]
```

```python
review_size
```

# Converting All Reviews Column into Numeric Value

```python
review_size['Reviews'] = pd.to_numeric(review_size['Reviews'], errors='coerce')
```

```python
# review_size['Size'] = pd.to_numeric(review_size['Size'], errors='coerce')
```

# Finding Values by Conditions in Reviews Column

```python
large_App_r = review_size[review_size['Reviews'] > 132]
large_App_r
```

# Finding Null Values in Reviews column of Category ART_AND_DESIGN

```python
nan_values_in_review = large_App_r['Reviews'].isna().sum()
nan_values_in_review
# large_App['Reviews'].dropna()
## large_App.dropna(subset=['Reviews'],inplace=True)
```

# Finding Values by Conditions in Size Column

```python
large_App_s = review_size[review_size['Size'] == '25M']
large_App_s
```

# Converting Size Column into Numeric Value

```python
review_size['Size'] = pd.to_numeric(review_size['Size'].str.replace('M',''), errors='coerce')
```

# Finding Value by Conditions in Size Column

```python
large_App_s = review_size[review_size['Size'] > 20]
large_App_s
```

# Finding Mean Value of Review Column

```python
mean_reviews = large_App_s['Reviews'].mean()
mean_reviews
```
