## Q1 How many apps are free in art and design

```python
ct=0
for i in df:
    if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
        ct+=1
print('There are ',ct,'Free Apps of ART_AND_DESIGN')
```

## Q2 How many apps are there with in Art_And_Design with Rating more than 4.5?

```python
ct=0
for i in df:
    if(i[1] == "ART_AND_DESIGN" and i[2] >= 4.5):
      ct+=1
print("There are",ct,"apps in ART_AND_DESIGN with Rating more than 4.5")
```
